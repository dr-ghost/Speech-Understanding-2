```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np

from background import *
```

```python
ind_dat = IndianLanguagesDataset(
    data_dir=r'Language Detection Dataset'
)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
model = LanguageClassifier(input_channels=100, num_classes=10)
```

```python
train_loss, val_acc, val_f1 = train(model, ind_dat, device=device, max_bins=100)
```

```python
#plot train loss
plt.figure(figsize=(10, 5))
plt.plot(np.arange(len(train_loss)), train_loss, label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs')
plt.legend()
plt.grid()
plt.show()
```

```python
#plotting validation accuracy
plt.figure(figsize=(10, 5))
plt.plot(np.arange(len(val_acc)), val_acc, label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Validation Accuracy over Epochs')
plt.legend()
plt.grid()
plt.show()
```

```python
#plotting validation f1 score
plt.figure(figsize=(10, 5))
plt.plot(np.arange(len(val_f1)), val_f1, label='Validation F1 Score')
plt.xlabel('Epochs')
plt.ylabel('F1 Score')
plt.title('Validation F1 Score over Epochs')
plt.legend()
plt.grid()
plt.show()
```
